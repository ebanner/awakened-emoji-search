```python
import open_clip

model, _, preprocess = open_clip.create_model_and_transforms('ViT-B-32', pretrained='laion2b_s34b_b79k')
model.eval()

tokenizer = open_clip.get_tokenizer('ViT-B-32')

tokenizer, model
```

```python
%image fatcat.png
```

```python
from PIL import Image

image = preprocess(Image.open("fatcat.png")).unsqueeze(0)

image
```

```python
from IPython.display import display

import torch

image = preprocess(Image.open('fatcat.png')).unsqueeze(0)
text = tokenizer(['fatcat'])

with torch.no_grad(), torch.autocast("cuda"):
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    fatcat_embedding = 0.5*image_features + 0.5*text_features
    fatcat_embedding /= fatcat_embedding.norm(dim=-1, keepdim=True)

    fatcat_embedding.cpu()

img = Image.open("fatcat.png")

print('fatcat')
display(img)

fatcat_embedding
```
